Ejercicio 2º Cread un subplot con dos gráficas, dónde en la primera tengamos el número de empleados por departamento y en la otra por tipo de puesto. Debéis:
Poner leyenda a la gráfica
Cada línea debe ir en un color
Ponerle nombre a los ejes y a la gráfica.
Poner marcadores a las lineas.

```python
df_ep = pd.read_csv("datos/empleados_puestos (1).csv")
df_ep.sample(3)
```

```python

df_empleados= df_ep["Departamento"].value_counts().reset_index()

df_empleados
```

```python
df_puesto= df_ep["Tipo Puesto"].value_counts().reset_index()

df_puesto
```

```python
fig, ax = plt.subplots(1,2, figsize = (30,10))
ax[0].plot(df_empleados["index"],df_empleados["Departamento"],color="green", marker = ".", linewidth=0.5,label ="empleados")
ax[1].plot(df_puesto["index"],df_puesto["Tipo Puesto"],color="pink", marker = "8", linewidth=0.5, label = "nª puesto")
ax[0].set_xticklabels(df_empleados["index"],rotation="vertical")

ax[0].set_title("Empleados departamento", fontsize=20)
ax[1]. set_title("Puesto empleados",fontsize=20)

ax[0].set_xlabel("puestos", fontsize=16)
ax[1].set_ylabel("números empleados", fontsize=16 )

ax[0].set_ylabel("números empleados",fontsize=16) 
ax[1].set_xlabel("tipo puesto",fontsize=16)

ax[0].legend(loc="upper right")
ax[1].legend(loc="upper right")
plt.show()
```
